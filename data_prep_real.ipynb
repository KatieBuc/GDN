```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/simriver/simriver.csv', index_col=0)
df = df[['t', 'locID', 'ind','y_obs']].pivot(index='t', columns='locID')
df
```

```python
df_y = df.y_obs.copy()
df_y['attack'] = df.ind.apply(lambda x: 0 if sum(x)==0 else 1, axis=1)
df_y.index.name = None
print(len(df_y[df_y.attack==1]))

df_y.index.name='timestamp'
df_y.columns.name=None
df_y
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(16, 6))
ax.plot(df_y)
plt.show()
```

```python
# save test data
df_y.columns = df_y.columns.astype(str)
df_y.columns = ['a'+i if i != 'attack' else i for i in df_y.columns]
df_y.iloc[:667,].to_csv('data/simriver/test.csv')
```

```python
# save train data
df_train = df_y.iloc[667:,:-1].reset_index(drop=True)
df_train.index.name='timestamp'
df_train.to_csv('data/simriver/train.csv')
```
